# ZFNet based on Zeiler and Fergus, 2013

ZFNet represents AlexNet with improved hyperparameters. Therefore, the basic architecture is similar to AlexNet. 
    A large part of this code is based on code from a fast.ai MOOC that will be publicly available in Jan 2019

## ZFNet

```python
#import torch module
import torch.nn as nn
```

```python
# ZFNet
class ZFNet(nn.Module):
    # num_classes = the number of classes that I would like to classify
    def __init__(self, num_classes=37):
        super(ZFNet,self).__init__()
        # conv layers
        self.feature = nn.Sequential(
            #first set
            nn.Conv2d(3, 96, kernel_size=7, stride=2, padding=1), #1
            nn.ReLU(inplace=True), #2
            nn.MaxPool2d(kernel_size=3, stride=2, padding=1), #3
            #second set
            nn.Conv2d(96, 256, kernel_size=5, stride=2), #4
            nn.ReLU(inplace=True), #5
            nn.MaxPool2d(kernel_size=3, stride=2, padding=1), #6
            #thrid set
            nn.Conv2d(256, 384, kernel_size=3, stride=1, padding=1), #7
            nn.ReLU(inplace=True), #8
            #fourth set
            nn.Conv2d(384, 384, kernel_size=3, stride=1, padding=1), #9
            nn.ReLU(inplace=True), #10
            #fifth set
            nn.Conv2d(384, 256, kernel_size=3, stride=1, padding=1), #11
            nn.ReLU(inplace=True), #12
            nn.MaxPool2d(kernel_size=3, stride=2), #13
        )
        # fully-connected layers
        self.classifier = nn.Sequential(
            nn.Dropout(),
            
            #sixth set
            nn.Linear(256*6*6, 4096),
            nn.ReLU(inplace=True),
            
            nn.Dropout(),
            # seventh set
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            
            # last layer
            nn.Linear(4096, num_classes),
        )
        
    def forward(self, x):
        x = self.feature(x)
        x = x.view(-1, 256 * 6 * 6) # to resize to match to matrix shape for the next linear layer
        x = self.classifier(x)
        return x
```

```python
# TO DO LIST - try own cross entorpy loss function
loss_func = nn.CrossEntropyLoss()
```

```python
def update(x,y,lr):
    # weight decay alpha
    wd = 1e-7
    # prediction
    y_hat = model(x)
    # weight decay
    w2 = 0.
    # p is parameter
    for p in model.parameters(): w2 += (p**2).sum()
    # add to regular loss
    loss = loss_func(y_hat, y) + w2*wd
    loss.backward()
    with torch.no_grad():
        for p in model.parameters():
            p.sub_(lr * p.grad)
            p.grad.zero_()
    return loss.item()
```

```python
losses = [update(x,y,lr) for x,y in data.train_dl]
```

```python
# loss doesn't look smooth because this dataset only has ~5000 images in a training set
plt.plot(losses);
```

```python
def update(x,y,lr):
    # use Adam optimiser
    opt = optim.Adam(model.parameters(), lr)
    y_hat = model(x)
    loss = loss_func(y_hat, y)
    loss.backward()
    # .step() does gradient descent
    opt.step()
    opt.zero_grad()
    return loss.item()
```
